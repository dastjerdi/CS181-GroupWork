```python
i = 0
def save(x):
    global i
    
    train_data = np.array([])
    p = pd.DataFrame(x)
    for rows in p.iterrows():
        index, chunk = rows
        chunk = np.array(chunk)
        print(i)
        row = np.array([])
        cstft = np.mean(lf.chroma_stft(np.array(chunk)[:-1]).T, axis=0)
        row = np.concatenate((row, cstft))
        spcent = np.mean(lf.spectral_centroid(np.array(chunk)[:-1]).T, axis=0)
        row = np.concatenate((row, spcent))
        mspec = np.mean(lf.melspectrogram(np.array(chunk)[:-1]).T, axis=0)
        row = np.concatenate((row, mspec))
        mfcc = np.mean(lf.mfcc(np.array(chunk)[:-1],n_mfcc=30).T, axis=0)
        row = np.concatenate((row,mfcc))
        tonnetz = np.mean(lf.tonnetz(np.array(chunk)[:-1]).T, axis=0)
        row = np.concatenate((row, tonnetz))
        rmse =  np.mean(lf.rmse(np.array(chunk)[:-1]).T, axis=0)
        row = np.concatenate((row, rmse))
        contrast =  np.mean(lf.spectral_contrast(np.array(chunk)[:-1]).T, axis=0)
        row = np.concatenate((row, contrast))
        row = np.append(row, np.array(chunk)[-1])
        i += 1

        np.append(train_data, row)
        print(len(row))
    return train_data

pool = mp.Pool(processes=8)
results = [pool.apply(save, args=(x,)) for x in train_data_reader]
```
