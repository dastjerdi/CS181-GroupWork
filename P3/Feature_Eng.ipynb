```python
%matplotlib inline
import re
import numpy as np
import pandas as pd
import string
import matplotlib.pyplot as plt
import librosa
import librosa.feature as lf
import librosa.display
import IPython.display as ipd
import numpy
import concurrent.futures
import multiprocessing as mp
```

```python
train_data_reader = pd.read_csv("train.csv", header=None, chunksize=100)
```

```python
i = 0
def EngF(x):
    global i
    
    train_data = []
    p = pd.DataFrame(x)
    for rows in p.iterrows():
        index, chunk = rows
        chunk = np.array(chunk)
        print(i)
        row = np.array([])
        cstft = np.mean(lf.chroma_stft(np.array(chunk)[:-1]).T, axis=0)
        row = np.concatenate((row, cstft))
        spcent = np.mean(lf.spectral_centroid(np.array(chunk)[:-1]).T, axis=0)
        row = np.concatenate((row, spcent))
        mspec = np.mean(lf.melspectrogram(np.array(chunk)[:-1]).T, axis=0)
        row = np.concatenate((row, mspec))
        mfcc = np.mean(lf.mfcc(np.array(chunk)[:-1],n_mfcc=30).T, axis=0)
        row = np.concatenate((row,mfcc))
        tonnetz = np.mean(lf.tonnetz(np.array(chunk)[:-1]).T, axis=0)
        row = np.concatenate((row, tonnetz))
        rmse =  np.mean(lf.rmse(np.array(chunk)[:-1]).T, axis=0)
        row = np.concatenate((row, rmse))
        contrast =  np.mean(lf.spectral_contrast(np.array(chunk)[:-1]).T, axis=0)
        row = np.concatenate((row, contrast))
        row = np.append(row, np.array(chunk)[-1])
        i += 1

        train_data.append([row])
    return train_data

pool = mp.Pool(processes=8)
results = [pool.apply(EngF, args=(x,)) for x in train_data_reader]
```

```python
import copy

other = copy.copy(results)
```

```python
flattened_features = [i[0] for group in results for i in group]
new_features = pd.DataFrame(flattened_features)
new_features.to_csv('Condensed_Set.csv')
```
